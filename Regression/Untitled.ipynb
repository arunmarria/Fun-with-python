```python
import pandas as pd
import numpy as np
```

```python
from  scipy.stats.distributions import bernoulli
```

```python
## biased coin

def coin_toss(n):
    dist = bernoulli(0.6)
    return dist.rvs(n)
```

```python
np.mean(coin_toss(15))
```

```python
## for large samples

rolling_mean = np.zeros((10000))

sample_size = np.arange(1, 10001,1)
for i in sample_size: 
    rolling_mean[i-1] = np.mean(coin_toss(i))
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(sample_size, rolling_mean)
plt.xscale('log')
```

As sample size increases the mean converges to the value p 

```python


def make_throws(number_of_samples, sample_size):
    start=np.zeros((number_of_samples, sample_size), dtype=int)
    for i in range(number_of_samples):
        start[i,:]=coin_toss(sample_size)
    return np.mean(start, axis=1)
    
```

```python

make_throws(number_of_samples=20, sample_size=10)
```

### A simple election model

```python
predictwise = pd.read_csv('predictwise.csv').set_index('States')
predictwise.head()
```

```python
def simulate_election(model, n_sim):
    simulations = np.random.uniform(size=(51, n_sim))
    obama_votes = (simulations < model.Obama.values.reshape(-1, 1)) * model.Votes.values.reshape(-1, 1)
    #summing over rows gives the total electoral votes for each simulation
    return obama_votes, obama_votes.sum(axis=0)
```
