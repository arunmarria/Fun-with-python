The notebook explores some basic data scrapping concepts. The exercises are referred from Havard's Data Science course CS109 - 2015(http://cs109.github.io/2015/index.html) 

```python
import urllib3
import numpy as np
```

```python
url = 'http://www.crummy.com/software/BeautifulSoup'
http_pool =  urllib3.PoolManager()
source_page= http_pool.urlopen('GET', url)
source_html = source_page.data.decode('utf-8')
print(source_html)
```

```python
##type of source_html
type(source_html)
## astype is string all functions on string are applicable
```

## Using Beautiful soup library

```python
import bs4 ## beautiful soup library
```

```python
bsoup = bs4.BeautifulSoup(source_html)

##print (bsoup)
print(bsoup.prettify())
```

```python
## trying to find link tags(a) without beautiful soup
import re ## regular expressions

re.findall('a', source_html)[1:10]

## finds all a's but we are interested only  the a tags which carry link 
```

```python
## using beautiful soup
print(bsoup.findAll('a'))
a_tags =  bsoup.find('a')
print(a_tags.get('href'))

```

```python
##getting all links
links= []
for tag in bsoup.findAll('a'):
    links.append(tag.get('href'))
    
links
    
```

```python
## filtering external links
ext_links =[]
for link in links:
    if link is not None:
        
        if link[:4] =='http':
            ext_links.append(link)
        
ext_links
```

## Parsing tree

```python

s = """<!DOCTYPE html><html><head><title>This is a title</title></head><body><h3> Test </h3><p>Hello world!</p></body></html>"""
## get bs4 object
tree = bs4.BeautifulSoup(s)

## get html root node
print(tree)
root_node = tree.html
print("root_nodez")
print(root_node)

## get head from root using contents
head = root_node.contents[0]

print(tree.title)
print(tree.body)

## get body from root
body = root_node.contents[1]

print(root_node.contents)
```

```python
## finding h3 tag


print(tree.body.h3)

print(root_node.contents[1].h3)

print(body.contents[0])
```

```python
##finding list of hall of fame entries from bodysoup page
list_hall =[]
for element in bsoup.ul.contents:
    if str(element)[:3]=='<li':
        list_hall.append(str(element.contents))
        
list_hall[0]
list_hall[1]
```

```python
##scrapping data science skills


url = 'http://www.indeed.com/jobs?q=data+scientist&l='
http  = urllib3.PoolManager()
url_page =  http.urlopen('GET', url)

##print(url_page)

source_url = url_page.data.decode('utf-8')


soup_html = bs4.BeautifulSoup(source_url)

soup_html
```

```python
## number of search results

job_count_string =  soup_html.find(id= 'searchCount').contents[0]
job_count_string = job_count_string.split()[-2]
print("Search yielded %s hits." % (job_count_string))
```

```python
## converting job_count_string into integers
job_count = int(''.join(job_count_string.split(',')))

print(job_count)
```

```python
## each page for result set has 10 ressults so number of pages to be scrapped.

num_pages = job_count//10 +1

print(num_pages)
```
